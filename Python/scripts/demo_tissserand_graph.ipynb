```python
import pykep as pk
from pykep.planet import jpl_lp
import numpy as np
```

## Determine the planets and the V-infinity ranges

```python
planets = [jpl_lp('venus'), jpl_lp('earth'), jpl_lp('mars'), jpl_lp('jupiter'), jpl_lp('saturn')]
mu = planets[0].mu_central_body
vinf_min = np.array([2, 3, 3, 2.5, 2]) * 1e3
vinf_max = np.array([8, 8, 8, 8, 8]) * 1e3
epoch = pk.epoch_from_string('2020-01-01 12:00:00.000')
safe_fb_rp = 1.1   # min flyby radius/planet radius
```

## Create Tisserand Graph

### Periapsis vs Energy

```python
import sys
sys.path.append("../src")
from tisserand_graph import tisserand_graph

tisserand_graph(planets, mu, epoch, vinf_min, vinf_max, safe_fb_rp=safe_fb_rp, xlabel="rp", ylabel="energy")
```

## Periapsis vs Apoapsis

```python
tisserand_graph(planets, mu, epoch, vinf_min, vinf_max, safe_fb_rp=safe_fb_rp, xlabel="rp", ylabel="ra",
                xlim=[0, 10], ylim=[0, 15])
```

## Periapsis vs Period

```python
tisserand_graph(planets, mu, epoch, vinf_min, vinf_max, safe_fb_rp=safe_fb_rp,
                xlabel="rp", ylabel="period", ylim=[0, 20*365])
```
